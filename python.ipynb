1.請將「imdb_top_1000.xlsx」檔案存到你的Google雲端硬碟中，並分享檔案權限的方式讀取檔案中『Data』工作表中的資料，並命名為imdb1000資料集。(10%)

```python
import pandas as pd

file_path = '/content/drive/MyDrive/Colab Notebooks/imdb_top_1000.xlsx'
imdb1000= pd.read_excel(file_path, sheet_name='Data')
imdb1000
```

2.請將imdb1000中欄位名稱為「Series_Title」改成「電影片名」；「Runtime」改為「影片長度」並重新存回imdb1000 (5%)。


```python
imdb1000.rename(columns={'Series_Title': '電影片名', 'Runtime': '影片長度'}, inplace=True)
imdb1000
```

3.請用info指令檢視資料集中變數是否符合「imdb_top_1000.xlsx」檔案中『變數定義』所認定之定義；若不是，請將該變數轉型成適當的資料型別(10%)。


```python
imdb1000['影片長度'] = imdb1000['影片長度'].astype(str).str.extract('(\d+)').astype(float)

imdb1000['IMDB_Rating'] = imdb1000['IMDB_Rating'].astype(float)
imdb1000['Meta_score'] = imdb1000['Meta_score'].astype(float)
imdb1000['No_of_Votes'] = imdb1000['No_of_Votes'].astype(int)

imdb1000['Gross'] = imdb1000['Gross'].astype(str).str.replace(',', '').astype(float)

imdb1000['Gross']
```

4.請將imdb1000中有遺失值的列刪除，另存成imdb_drop(5%)。

```python
imdb_drop = imdb1000.dropna()
imdb_drop
```

5.請用「if…elif…elif…else…判斷式」將imdb_drop中「IMDB_Rating」重新定義一個變數為『評等』存回imdb_drop這個資料集內，變數定義如下：(10%)

```python
def define_rating(row):
    if row['IMDB_Rating'] >= 9.0:
        return 'A'
    elif 8.5 <= row['IMDB_Rating'] < 9.0:
        return 'B'
    elif 8.0 <= row['IMDB_Rating'] < 8.5:
        return 'C'
    else:
        return 'D'

imdb_drop['評等'] = imdb_drop.apply(define_rating, axis=1)
```

6.請將imdb_drop中的「電影片名」定義成index並重新存回imdb_drop (5%)。

```python
imdb_drop.set_index('電影片名', inplace=True)
```

7.請將imdb_drop中「影片長度」、「IMDB_Rating」、「Meta_score」、「No_of_Votes」、「Gross」5個欄位資料另存為imdb_numeric資料集(5%)。


```python
imdb_numeric = imdb_drop[['影片長度', 'IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross']]
imdb_drop['Gross']
```

8.計算imdb_numeric所有欄位的平均數、標準差、最大值、最小值，以及變數間的共變異數及相關係數矩陣(10%)。

```python
mean_values = imdb_numeric.mean()
std_dev = imdb_numeric.std()
max_values = imdb_numeric.max()
min_values = imdb_numeric.min()

cov_matrix = imdb_numeric.cov()
corr_matrix = imdb_numeric.corr()

print(mean_values)
print(std_dev)
print(max_values)
print(min_values)
print(cov_matrix)
print(corr_matrix)
```

13.請篩選imdb_drop中「Certificate」為U的項目，且欄位保留「Certificate」及「Meta_Score」2個變數，另存成imdb_U資料集(5%)。


```python
imdb_U = imdb_drop[imdb_drop['Certificate'] == 'U'][['Certificate', 'Meta_score']]
```

14.請繪製imdb_U中「Meta_Score」變數的莖葉圖(5%)。

```python
import pandas as pd
import matplotlib.pyplot as plt

stem_and_leaf = {}
for score in imdb_U['Meta_score']:
    tens_digit = score // 10
    ones_digit = score % 10
    if tens_digit not in stem_and_leaf:
        stem_and_leaf[tens_digit] = []
    stem_and_leaf[tens_digit].append(ones_digit)

for tens_digit, leaves in sorted(stem_and_leaf.items()):
    print(f"{tens_digit} | {' '.join(map(str, sorted(leaves)))}")

plt.figure(figsize=(6, 4))
for tens_digit, leaves in sorted(stem_and_leaf.items()):
    plt.scatter([tens_digit] * len(leaves), leaves, marker='o', color='b')
plt.xlabel("Tens Digit")
plt.ylabel("Leaves")
plt.title("Stem-and-Leaf Plot for Meta_Score in imdb_U")
plt.grid(True)
plt.show()
```
